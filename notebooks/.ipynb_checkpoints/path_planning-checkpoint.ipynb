```python
circle_balises = cv2.imread('../assets/additional/path_planning.png')
```

```python
aruco_info = json.load(open('../assets/aruco/state.json', 'r'))
calibration_path = '../assets/additional/calibration_test'
detector = VisionDetector(circle_balises, (1100, 700), calibration_path, **aruco_info)
img_corrected = detector.get_corrected_image(circle_balises, hide_aruco=False, undistort=False)
plt.imshow(cv2.cvtColor(img_corrected, cv2.COLOR_BGR2RGB))
```

```python
from src.vision import extend_polygons
from src.utils import add_circles, add_polygons
```

```python
circles = detector.find_circles(img_corrected, 1, 300, 300, 30)
polygons = detector.find_polygons(img_corrected)
processed_image = add_circles(img_corrected, circles)
processed_image = add_polygons(processed_image, polygons)
plt.imshow(cv2.cvtColor(processed_image, cv2.COLOR_BGR2RGB))
```

## Path planning

```python
from src.path_planing import create_adjacency_matrix, create_obstacle_map, path_planning
from src.utils import add_graph, add_route
```

```python
robot_size = 100
polygons, circles = detector.find_shapes(img_corrected, border_size=10)
extended_polygons = extend_polygons(polygons, robot_size)

poly_image = add_polygons(img_corrected, extended_polygons)
circle_image = add_circles(poly_image, circles)
plt.imshow(cv2.cvtColor(circle_image, cv2.COLOR_BGR2RGB))
```

```python
obstacle_map = create_obstacle_map(extended_polygons, img_corrected.shape)
adjacency_matrix, points = create_adjacency_matrix(extended_polygons, obstacle_map)
graph_image = add_graph(circle_image, points, adjacency_matrix, color=(255, 0, 0))
plt.imshow(cv2.cvtColor(graph_image, cv2.COLOR_BGR2RGB))
```

```python
polygons = detector.find_polygons(img_corrected)
extended_polygons = extend_polygons(polygons, robot_size)
robot_position, _, _ = detector.get_robot_pose(img_corrected, is_corrected=True)
circles = detector.find_circles(img_corrected, 1, 300, 300, 30)
circle_positions = circles.squeeze()[:, :2]

optimal_route, optimal_route_positions, _, _ = path_planning(robot_position, circle_positions, extended_polygons, processed_image.shape)
```

```python
route_image = add_route(circle_image.copy(), np.int32(optimal_route_positions))
plt.imshow(cv2.cvtColor(route_image, cv2.COLOR_BGR2RGB))
```
