```python
#===============ROBOT INSTRUCTIONS==================================
#function to get the value of the prox
async def get_prox_value():
    await node.wait_for_variables({"prox.horizontal"})
    #await client.sleep(0.1)
    return list(node["prox.horizontal"])

#set the speed of the motors
async def set_motor_speed(left, right):
    await node.set_variables({"motor.left.target": [left], "motor.right.target": [right]})
```

```python
import math
from src.path_planning import path_planning
from src.vision import extend_polygons
from src.utils import add_robot, add_route, add_circles, add_polygons, add_point, add_line
from src.geometry import convert_points, convert_vec 
from src.navigation import change_target, FSM, global_control, global_control_path

aw(node.set_variables({"leds.top" : leds_top}))
vec_to_print = np.array([0.,0.])

while True:
    
    #get the prox sensors values
    prox = await get_prox_value()
    ret, frame = cam.read()

    # TODO: fix undistort
    corrected_image = detector.get_corrected_image(frame)

    #stop the program if the robot is not on the map (make the kidnapping)
    try:
        robot_pos_cv, robot_orientation_cv, robot_corners_cv = detector.get_robot_pose(corrected_image, is_corrected=True)
    
    except AttributeError as e:
        print(f"Erreur dtecte : {e}")
        break

    # robot_pos_cv, robot_orientation_cv, robot_corners_cv = detector.get_robot_pose(corrected_image, is_corrected=True)
    
    #display the image
    #img_polygons = add_polygons(corrected_image, extended_polygons)
    #img_circles = add_circles(img_polygons, circles)
    #img_route = add_route(img_circles, np.int32(optimal_route_positions))
    img_aruco = add_robot(corrected_image, robot_pos_cv, np.array(robot_orientation_cv), np.array(robot_corners_cv))
    img_line = add_line(img_aruco, path_cv["v_dir"],path_cv["point"])
    img_final = add_point(img_line, convert_points(current_target, y_max))
    

    #convert the working point and vector into mathematical coordinate system
    robot_pos_m = convert_points(robot_pos_cv, y_max)
    robot_orientation_m = convert_vec(robot_orientation_cv)
#======== ROBOT MOVEMENTS AND CALCULATION GLOBAL NAVIGATION========================================

    
    #controller = global_control_path(path_m, robot_pos_m, robot_orientation_m, 0.1)
    

    
    

#================================= FONCTIONNE =======================================
    state, leds_top, speed_l, speed_r = FSM(state, leds_top, prox, path_m, robot_pos_m, robot_orientation_m)
    # if state==1:
    #     await set_motor_speed(0, 0)
    #     break
   
    await set_motor_speed(speed_l, speed_r)
    aw(node.set_variables({"leds.top" : leds_top}))

    #look if target is reached, and change it if so
    stop, objectives_m, global_path_m = change_target(robot_pos_m, current_target, global_path_m, path_m, y_max, path_cv, objectives_m)
    if stop: break #all target are reached so end the program
    

    cv2.imshow('detected image', img_final)


    print(global_path_m)
    key = cv2.waitKey(1)
    # Press 'space' to exit the loop
    if key == ord(' '):
        print("programme en pause")
        await set_motor_speed(0, 0)
        while(1):
            if cv2.waitKey(1) == ord(' '):
                break
                
    # Press 'q' to exit the loop
    if key == ord('q'):
        break

    #print('i : ', i)
    sleep(dt/1000)

# Release the capture and writer objects
await set_motor_speed(0, 0)

#cam.release()
cv2.destroyAllWindows()
```
